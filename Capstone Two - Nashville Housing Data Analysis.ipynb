```python
corr = housing.corr()
fig, ax = plt.subplots(figsize=(12,6)) 
sns.heatmap(corr, annot=True,linewidths=.1,
            xticklabels=corr.columns.values,
            yticklabels=corr.columns.values,
            ax=ax)
plt.title("Heatmap of Housing Variable Correlations")
plt.show()
```

```python
corr = corr.sort_values("Sale Price", ascending=False)
corr['Sale Price'].plot(kind='bar')
plt.title("Variables Correlated with Sale Price")
plt.ylabel("Correlation Strength")
plt.show()
```

```python
import statsmodels.api as sm

dependentVar = housing['Sale Price']
filt = ['Total Value', 'Building Value', 'Land Value', 'Finished Area', 'Full Bath', 
        'Bedrooms', 'Acreage', 'Year Built']
#----write your code below------
housing = housing[filt]
housing['Sale Price'] = dependentVar
housing = housing.fillna(0)
X = housing.drop(['Sale Price'],axis=1)
X = sm.add_constant(X)
y = housing['Sale Price']
OLSmodel = sm.OLS(y, X)
OLSmodelResult = OLSmodel.fit()
OLSmodelResult.summary()
```

```python

plt.rcParams['figure.figsize'] = (15,5)
#----write your code below------
housing['Sale Price'] = housing['Sale Price']
ax = OLSmodelResult.predict(X).plot(linewidth=3, marker='.')
ax2 = ax.twinx()
ax2.plot(housing.SalePrice, 'mediumseagreen', linewidth=3, marker='*')
#ax.legend(bbox_to_anchor=(1.04,1), loc="upper left")
plt.tight_layout()
plt.title("Regression Plot vs. Actual Sale Prices")
plt.show()
```
